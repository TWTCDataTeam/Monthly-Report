```python
import pandas as pd
import glob
import re
from datetime import datetime
from tqdm import tqdm
import sys
from mof_data_collector import *
from Industry_processer import *
from Table_processer import *
from config import FILE_OUTPUT_PATH
# suppress scientific notation
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
mof_data = mof_txt_df()
```

```python
ind_define_dict = get_market_sec_industry_define_dict()
ind_define_dict
```

### 設定主要參數

```python
latest_year = int(sys.argv[1])
latest_month = int(sys.argv[2])
year_list = [e for e in range(2013,latest_year)]
print(year_list)
country_list = ["中國大陸","香港"]
industry_list = list(ind_define_dict.keys())
```

### 計算年資料

```python

output_table = pd.DataFrame(industry_list, columns=["Industry"])
for i in range(0,len(year_list)):
    if year_list[i] >= 2019:
        print("{} started".format(year_list[i]))
        tmp = caculate_txt_year_data(year_list[i],mof_data,country_list, ind_define_dict)
        output_table = output_table.merge(tmp, on="Industry", how="outer")
        print("===== merge finished =====")
        
    else :
        print("{} started".format(year_list[i]))
        tmp = caculate_year_data(year_list[i], country_list, ind_define_dict)
        #print(tmp)
        output_table = output_table.merge(tmp, on="Industry", how="outer")
        print("===== merge finished =====")

output_table
```

# 計算最新年度資料

```python
output_latestyear_table = caculate_latest_txt_year_data(latest_year, latest_month, mof_data, country_list,ind_define_dict)
output_latestyear_table
```

# 數據合併

```python
output = output_table.merge(output_latestyear_table,on = "Industry", how="outer")
output
```
